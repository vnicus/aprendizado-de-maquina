# Regressão Linear Multipla

```python
import numpy as np

# 1 Lendo os dados com NumPy
# Pulamos o Cabeçalho (skip_header=1)
# Extraimos apenas as colunas da interesse para evitar erro com strings:
# 2: price (Y), 3: bedrooms (X1), 4: bathrooms (X2), 5: sqft_living (X3)
dados = np.genfromtxt('kc_house_data.csv', delimiter=',', skip_header=1, usecols=(2,3,4,5))

# 2. Separando a variavel alvo (Y) e as variaveis preditoras (X) 
y = dados[:, 0] # O preço é a primeira coluna do nosso recorte
X_features = dados[:, 1:] # O restate são as features

# 3. Adicionando o termo de Intercepto (Viés/Bias)
# Precisamos adicionar uma coluna de número "1" no inicio da matriz X
# Isso representa o beta_0 na equação (o valor base da casa se tudo fosse zero)
m = X_features.shape[0]
X = np.c_[np.ones(m), X_features]

# 4. Aplicando a Equação Normal: Beta = (X.T * X)^-1 * X.T * Y
X_T = X.T # Transporta de X
# np.linalg.inv calcula a inversa da matriz
# o método .dot() ou operador @ realizar a multiplicação de matrizes
beta = np.linalg.inv(X_T @ X) @ X_T @ y

print('=== Coeficientes Descobertos ===')
print(f'Intercepto (Beta 0): {beta[0]:.2f}')
print(f'Peso para Quartos (Beta 1): {beta[1]:.2f}')
print(f'Peso para Banheiros (Beta 2): {beta[2]:.2f}')
print(f'Peso para Área/sqft (Beta 3): {beta[3]:.2f}\n')

# 5. Fazendo um previsão
# Vamos prever o valor da primeira casa do dataset: 3 quartos, 1 banheiro, 1180 sqft
casa_teste = np.array([1, 3, 1, 1180]) # Lembre-se do "1" inicial para o intercepto!
preco_previsto = casa_teste @ beta

print('=== Teste de Previsão ===')
print(f'Preco Real da Casa 1: $ {y[0]:.2f}')
print(f'Preco PREVISTO pelo nosso modelo: $ {preco_previsto:.2f}')
print(f"Diferença ${abs(y[0] - preco_previsto):.2f} (Erro de Previsão)")
```

Utilizando panda e mais variaveis indepentes

```python
import pandas as pd
import numpy as np

# 1. Lendo os dados com Pandas
# Pulamos o cabeçalho (skip_header=1)
# Extraimos apenas as colunas de interesse para evitar erro com string:
# 2: price (Y), 3: bedrooms (X1), 4: bathrooms (X2), 5: sqft_living (X3)

# Usando pandas
df = pd.read_csv('kc_house_data.csv', skipinitialspace=True)

# Os nomes ainda podem ter espaços no final, então normalize tudo:
df.columns = df.columns.str.strip()

# 2. Separando a variável alvo (Y) e as variáveis preditoras (X)

# Pandas
y = df['price']
X_features = df[['bedrooms', 'bathrooms', 'sqft_lot', 'floors', 'condition', 'yr_built', 'yr_renovated']]

# 3. Adicionando o termo de Intercepto (Viés/Bias)
# Precisamos adicionar uma coluna de números "1" no início da matriz X
# Isso represanta o beta_0 na equação (o valor base da casa se tudo fosse zero)
m = X_features.shape[0] # Número de linhas (exemplo)
X = np.c_[np.ones(m), X_features]

# 4. Aplicando a Equação Normal: Beta = (X.T * T)^-1 * X.T * Y
X_T = X.T # Transposta de X
# np.linealg.inv calcula a matriz inversa da matriz
# O método .dot() ou o operador @ realiza a multiplicação de matrizes
beta = np.linalg.inv(X_T @ X) @ X_T @ y

print("=== Coeficientes Descobertos ===")
print(f"Intercepto (Beta 0): {beta[0]:.2f}")
print(f"Peso para Quartos (Beta 1): {beta[1]:.2f}")
print(f"Peso para Banheiros (Beta 2): {beta[2]:.2f}")
print(f"Peso para Área/sqft (Beta 3): {beta[3]:.2f}")
print(f"Peso para Floors (Beta 4): {beta[4]:.2f}\n")


# 5. Fazendo uma previsão 
casa_teste = np.array([1, 3, 1, 5650, 1, 3, 1955, 0]) # Lembre-se do "1" inicial para o intercepto!
preco_previsto = casa_teste @ beta

print("=== Teste de Previsão ===")
print(f"Preço REAL da casa 1: $ {y[0]:.2f}")
print(f"Preço PREVISTO pelo nosso modelo: $ {preco_previsto:.2f}")
print(f"Diferença ${abs(y[0] - preco_previsto):.2f} (Erro de Previsão)")

casa2 = np.array([1, 3, 2, 7242, 2, 3, 1951, 1991])
preco2 = casa2 @ beta

print("=== Teste de Previsão 2 ===")
print(f"Preço REAL da casa 2: $ {y[1]:.2f}")
print(f"Preço PREVISTO pelo nosso modelo: $ {preco2:.2f}")
print(f"Diferença ${abs(y[1] - preco2):.2f} (Erro de Previsão)")
```
